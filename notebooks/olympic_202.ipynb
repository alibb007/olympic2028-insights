## Something's Off with the Medal Counts!
### The Issue:
Our data overestimates medals because it counts multiple rows for each team member, inflating the totals.

### The Fix:
Using the `event_type` column, we can correct this. By treating 'TEAM', 'HCOUP', and 'COUP' as **True** (team events) and all others as **False** (individual events), we ensure that team medals are counted as **one**—just like in the Olympics.


```python
df_merge['event_type'].value_counts()
```

```python
# Create a condition to identify team events using the 'event_type' column
team_event_types = ['TEAM', 'HCOUP', 'COUP']

# Mark team events as True or False based on the 'event_type' column in df_merge
df_merge['is_team_event'] = df_merge['event_type'].isin(team_event_types)
df_merge.head(10)
```

```python
# Separate team and individual events
team_events = df_merge[df_merge['is_team_event']].drop_duplicates(subset=['event', 'medal_type', 'country_code'])
individual_events = df_merge[~df_merge['is_team_event']]

# Adjusted counting for team events: count only one medal per event
team_medals = team_events.groupby(['Country Name', 'NOC', 'country_code', 'medal_type']).size().unstack(fill_value=0)
team_medals.columns = ['Total_Gold', 'Total_Silver', 'Total_Bronze']  # Reordered columns
team_medals = team_medals.reset_index()

# Counting for individual events
individual_medals = individual_events.groupby(['Country Name', 'NOC', 'country_code', 'medal_type']).size().unstack(fill_value=0)
individual_medals.columns = ['Total_Gold', 'Total_Silver', 'Total_Bronze']  # Reordered columns
individual_medals = individual_medals.reset_index()

# Combine both team and individual medals
combined_medals = pd.concat([team_medals, individual_medals], ignore_index=True)

# Aggregate the combined results to get the final medal counts
grouped_df = combined_medals.groupby(['Country Name', 'NOC', 'country_code']).sum().reset_index()

# Calculate Total_Medals from the sum of Total_Gold, Total_Silver, and Total_Bronze
grouped_df['Total_Medals'] = grouped_df[['Total_Gold', 'Total_Silver', 'Total_Bronze']].sum(axis=1)

# Merge this with the original 'Total' column to compare
comparison_df = grouped_df.merge(df_merge[['Country Name', 'Total']].drop_duplicates(), on='Country Name')

# Display the comparison to check if the numbers match
comparison_df
```

# Clustring

```python
from sklearn.preprocessing import StandardScaler

# Selecting the relevant columns
features = df_reordered[['Gold', 'Silver', 'Bronze', 'Total']]

# Scaling the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans

# Initialize the KMeans algorithm
kmeans = KMeans(n_clusters=4, random_state=42)

# Fit the model
kmeans.fit(scaled_features)

# Get the cluster labels
df_reordered['Cluster'] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Assuming scaled_features is your data after scaling

# Step 1: Calculate WCSS for different numbers of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Step 2: Manually Set the Best Number of Clusters to 3
best_cluster_number = 3  # Manually setting the best number of clusters to 3

# Step 3: Plotting the Elbow Graph with 3 Clusters Highlighted
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')

# Highlight the elbow point at 3 clusters
plt.axvline(x=best_cluster_number, color='red', linestyle='--')
plt.scatter(best_cluster_number, wcss[best_cluster_number - 1], color='red', s=100)  # Mark the point at 3 clusters

# Annotate the chosen number of clusters
plt.text(best_cluster_number, wcss[best_cluster_number - 1], f' Best: {best_cluster_number}', color='red', fontsize=12, ha='left')

plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd

# Assuming df_reordered is your original DataFrame

# Step 1: Prepare the Data and Fix the Order
df_reordered_sorted = df_reordered.sort_values('Total', ascending=False)
df_reordered_sorted = df_reordered_sorted[['Country Name', 'Gold', 'Silver', 'Bronze', 'Total']].drop_duplicates().dropna()

# Keep the order fixed
country_order = df_reordered_sorted['Country Name'].tolist()

# Step 2: Perform Hierarchical Clustering for Total Medals
linked_total = linkage(df_reordered_sorted[['Total']], method='ward')

# Step 3: Generate the Dendrogram for Total Medals
plt.figure(figsize=(20, 7))
dendrogram(
    linked_total,
    labels=country_order,  # Use the fixed order
    leaf_rotation=90,
    leaf_font_size=12,
    color_threshold=0.7 * max(linked_total[:, 2]),
)
plt.title('Dendrogram of Countries based on Total Medals')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.show()

# Step 4: Perform Hierarchical Clustering for Gold Medals
linked_gold = linkage(df_reordered_sorted[['Gold']], method='ward')

# Step 5: Generate the Dendrogram for Gold Medals
plt.figure(figsize=(20, 7))
dendrogram(
    linked_gold,
    labels=country_order,  # Use the fixed order
    leaf_rotation=90,
    leaf_font_size=12,
    color_threshold=0.7 * max(linked_gold[:, 2]),
)
plt.title('Dendrogram of Countries based on Gold Medals')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.show()

# Step 6: Perform Hierarchical Clustering for Silver Medals
linked_silver = linkage(df_reordered_sorted[['Silver']], method='ward')

# Step 7: Generate the Dendrogram for Silver Medals
plt.figure(figsize=(20, 7))
dendrogram(
    linked_silver,
    labels=country_order,  # Use the fixed order
    leaf_rotation=90,
    leaf_font_size=12,
    color_threshold=0.7 * max(linked_silver[:, 2]),
)
plt.title('Dendrogram of Countries based on Silver Medals')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.show()

# Step 8: Perform Hierarchical Clustering for Bronze Medals
linked_bronze = linkage(df_reordered_sorted[['Bronze']], method='ward')

# Step 9: Generate the Dendrogram for Bronze Medals
plt.figure(figsize=(20, 7))
dendrogram(
    linked_bronze,
    labels=country_order,  # Use the fixed order
    leaf_rotation=90,
    leaf_font_size=12,
    color_threshold=0.7 * max(linked_bronze[:, 2]),
)
plt.title('Dendrogram of Countries based on Bronze Medals')
plt.xlabel('Country')
plt.ylabel('Distance')
plt.show()
```
